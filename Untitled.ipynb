```python
import holidays
from pandas.tseries.holiday import Holiday, AbstractHolidayCalendar, nearest_workday
from pandas.tseries.offsets import CustomBusinessDay
import pandas as pd

# Define the Pakistan holiday calendar
class PakistanHolidayCalendar(AbstractHolidayCalendar):
    rules = [
        Holiday("Pakistan Day", month=3, day=23, observance=nearest_workday),
        Holiday("Labour Day", month=5, day=1, observance=nearest_workday),
        Holiday("Independence Day", month=8, day=14, observance=nearest_workday),
        Holiday("Iqbal Day", month=11, day=9, observance=nearest_workday),
        Holiday("Quaid-e-Azam Day", month=12, day=25, observance=nearest_workday),
    ]

# Create a CustomBusinessDay with the Pakistan Holiday Calendar
pkb = CustomBusinessDay(calendar=PakistanHolidayCalendar())

# Create a list of business days excluding Pakistan holidays for December 2024
business_days = pd.date_range(start='12/1/2024', end='12/30/2024', freq=pkb)

print("Business days in December 2024 excluding Pakistan holidays:")
print(business_days)

# Create a date range for July 2024 using the custom business day frequency
rng = pd.date_range(start='7/1/2024', end='7/31/2024', freq=pkb)

print("Business days in July 2024 excluding Pakistan holidays:")
print(rng)
```
